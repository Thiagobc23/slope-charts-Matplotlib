```python
countries = ['Ireland', 'Norway',
             'Switzerland', 'Cayman Islands',
             'China, Macao Special Administrative Region']

fig, ax = plt.subplots(1, figsize=(10,10))

for i in countries:
    # get a single country from the list
    temp = df[df['Country or Area'] == i]
    # plot the lines
    plt.plot(temp.Year, temp.Value, marker='o', markersize=5)
    # replace large labels
    if i == 'China, Macao Special Administrative Region':
        i = 'Macao (China)'
    # end label
    plt.text(temp.Year.values[0]+0.02, temp.Value.values[0], i)
    # start label
    plt.text(temp.Year.values[1]-0.02, temp.Value.values[1], i, ha='right')
    
# x limits, x ticks, and y label 
plt.xlim(2017.5,2019.5)
plt.xticks([2018, 2019])

# get y ticks, replace 1,000 with k, and draw the ticks
yticks = plt.yticks()[0]
plt.yticks(yticks, ['{}k'.format(i/1000) for i in yticks])

plt.show()
```

```python
countries = ['Ireland', 'Norway',
             'Switzerland', 'Cayman Islands',
             'China, Macao Special Administrative Region']

colors = ['#769465', '#69BDE0', '#E06D5E', '#2F7694', '#94E069']

fig, ax = plt.subplots(1, figsize=(10,10), facecolor='darkgrey')
ax.set_facecolor('darkgrey')
for i, v in enumerate(countries):
    # get a single country from the list
    temp = df[df['Country or Area'] == v]
    # plot the lines
    plt.plot(temp.Year, temp.Value, color=colors[i], lw=2.5, marker='o', markersize=5)
    # replace large labels
    if v == 'China, Macao Special Administrative Region':
        v = 'Macao (China)'
    # end label
    plt.text(temp.Year.values[0]+0.02, temp.Value.values[0], v)
    # start label
    plt.text(temp.Year.values[1]-0.02, temp.Value.values[1], v, ha='right')
    
# x limits, x ticks, and y label 
plt.xlim(2017.5,2019.5)
plt.xticks([2018, 2019])
plt.ylabel('USD')
# get y ticks, replace 1,000 with k, and draw the ticks
yticks = plt.yticks()[0]
plt.yticks(yticks[1:-1], ['{}k'.format(i/1000) for i in yticks[1:-1]])

# grid
ax.xaxis.grid(color='black', linestyle='solid', which='both', alpha=0.9)
ax.yaxis.grid(color='black', linestyle='dashed', which='both', alpha=0.1)

# remove spines
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['top'].set_visible(False)

plt.title('GDP Per Capta\n', loc='left', fontsize=20)
plt.show()
```

```python
countries = ['Ireland', 'Norway',
             'Switzerland', 'Cayman Islands',
             'China, Macao Special Administrative Region']

colors = ['#769465', '#69BDE0', '#E06D5E', '#2F7694', '#94E069']

fig, ax = plt.subplots(1, figsize=(10,10), facecolor='darkgrey')
ax.set_facecolor('darkgrey')
for i, v in enumerate(countries):
    temp = df[df['Country or Area'] == v]
    plt.plot(temp.Year, temp.Value, color=colors[i], lw=2.5)
            
    plt.text(temp.Year.values[0]+0.02, 
             temp.Value.values[0], 
             '{:,.2f}'.format(temp.Value.values[0]))
    
    plt.text(temp.Year.values[1]-0.02, 
             temp.Value.values[1], 
             '{:,.2f}'.format(temp.Value.values[1]), va='center', ha='right')
    

plt.xlim(2017.5,2019.5)
plt.xticks([2018, 2019])
yticks = plt.yticks()[0]
plt.yticks([])

# grid
ax.xaxis.grid(color='black', linestyle='solid', which='both', alpha=0.9)

# remove spines
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['top'].set_visible(False)

plt.title('GDP Per Capta\n', loc='left', fontsize=20)

for i, v in enumerate(countries):
    if v == 'China, Macao Special Administrative Region':
        countries[i] = 'Macao (China)'

plt.legend(countries, loc='upper left', frameon=False)
plt.show()
```

```python
import numpy as np

countries = ['Ireland', 'Norway',
             'Switzerland', 'Cayman Islands',
             'China, Macao Special Administrative Region']

colors = ['#769465', '#69BDE0', '#E06D5E', '#2F7694', '#94E069']

fig, ax = plt.subplots(1, figsize=(10,10), facecolor='darkgrey')
ax.set_facecolor('darkgrey')

temp = df[df['Country or Area'].isin(countries)]
temp2018 = temp[temp['Year'] == 2018]
temp2019 = temp[temp['Year'] == 2019]
y = np.arange(0, len(countries))
plt.barh(y+0.15, temp2018.Value, height=0.3, color='#477998')
plt.barh(y-0.15, temp2019.Value, height=0.3, color='#FFBF00')

for i, v in enumerate(temp2018.Value):
    plt.text(v+500, y[i]+0.15, '{:,.2f}'.format(v), va='center')
for i, v in enumerate(temp2019.Value):
    plt.text(v+500, y[i]-0.15, '{:,.2f}'.format(v), va='center')


countries = temp['Country or Area'].unique().tolist()
for i, v in enumerate(countries):
    if v == 'China, Macao Special Administrative Region':
        countries[i] = 'Macao (China)'

plt.yticks(y, countries)

# remove spines
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['top'].set_visible(False)

# grid
ax.set_axisbelow(True)
ax.xaxis.grid(color='black', linestyle='dashed', which='both', alpha=0.3)

plt.show()
```

```python
import numpy as np

countries = ['Ireland', 'Norway',
             'Switzerland', 'Cayman Islands',
             'China, Macao Special Administrative Region']

colors = ['#769465', '#69BDE0', '#E06D5E', '#2F7694', '#94E069']

fig, ax = plt.subplots(1, figsize=(10,10), facecolor='darkgrey')
ax.set_facecolor('darkgrey')

for i, c in enumerate(countries):
    temp = df[df['Country or Area'] == c] 
    color = '#119822'
    marker = "^"
    if temp.Value.values[0] < temp.Value.values[1]: 
        color = '#D64933'
        marker = "v"
    plt.plot([i,i], [temp.Value.values[0], temp.Value.values[1]], marker=marker, markersize=8, color=color)
    if c == 'China, Macao Special Administrative Region':
        countries[i] = 'Macao (China)'
    
# x limits, x ticks, and y label
plt.xticks(np.arange(0,len(countries)), countries, rotation=45, ha='right')
plt.ylabel('USD')
# get y ticks, replace 1,000 with k, and draw the ticks
yticks = plt.yticks()[0]
plt.yticks(yticks[1:-1], ['{}k'.format(i/1000) for i in yticks[1:-1]])

# grid
ax.yaxis.grid(color='black', linestyle='dashed', which='both', alpha=0.2)

# remove spines
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['top'].set_visible(False)
```
